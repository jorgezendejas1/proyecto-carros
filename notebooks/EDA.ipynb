```python
import pandas as pd
import plotly.express as px

# Leer los datos
df = pd.read_csv('../vehicles_us.csv')

# Crear un histograma rápido
fig_hist = px.histogram(df, x="type")
fig_hist.show()

# Crear un gráfico de dispersión rápido
fig_scatter = px.scatter(df, x="odometer", y="price")
fig_scatter.show()
```
